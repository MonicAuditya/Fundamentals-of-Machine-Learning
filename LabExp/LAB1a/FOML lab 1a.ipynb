```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the dataset
data = pd.read_csv('iris.csv')
```

```python
# Display the first few rows of the dataset
data.head(150)
```

```python
data.shape
```

```python
data.loc[data['variety'] == 'Setosa']
```

```python
data_setosa = data.loc[data['variety'] == 'Setosa']
data_versicolor = data.loc[data['variety'] == 'Versicolor']
data_virginica = data.loc[data['variety'] == 'Virginica']
```

```python
plt.scatter(data_setosa['sepal.width'], np.zeros_like(data_setosa['sepal.width']))
plt.scatter(data_versicolor['sepal.width'], np.zeros_like(data_versicolor['sepal.width']))
plt.scatter(data_virginica['sepal.width'], np.zeros_like(data_virginica['sepal.width']))
plt.xlabel('Sepal Width')
plt.ylabel('Species')
plt.show()
```

```python
plt.scatter(data_setosa['sepal.length'], np.zeros_like(data_setosa['sepal.length']))
plt.scatter(data_versicolor['sepal.length'], np.zeros_like(data_versicolor['sepal.length']))
plt.scatter(data_virginica['sepal.length'], np.zeros_like(data_virginica['sepal.length']))
plt.xlabel('Sepal length')
plt.ylabel('Species')
plt.show()
```

```python
sns.FacetGrid(data, hue="variety", height=5).map(plt.scatter, "sepal.width", "petal.width").add_legend()
```

```python
sns.pairplot(data, hue="variety", height=2)
```
